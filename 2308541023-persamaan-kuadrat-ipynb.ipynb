```python
import numpy as np
import matplotlib.pyplot as plt

def persamaan_kuadrat_dari_akar (a,b,c): 
    # menghitung koefisien dari persamaan kuadrat
    a = 2
    b = 3
    c = 5
    
    # mengembalikan persamaan kuadrat dalam bentuk ax^2 + bx + c = 0
    return f"x^2 + {b}x + {c} = 0"

def plot_fungsi_kuadrat (a,b,c) :
    # menghitung koefisien dari persamaan kuadrat
    a = 2
    b = 3
    c = 5
    
    # menghasilkan nilai x
    x = np.linspace(-10, 10, 400)
    
    # menghitung nilai y
    y = a * x**2 + b * x + c
    
    # membuat plot grafik
    plt.plot(x, y)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Grafik Fungsi Kuadrat")
    plt.grid(True)
    plt.show()
persamaan = persamaan_kuadrat_dari_akar (a,b,c)
print(persamaan)

plot_fungsi_kuadrat (a,b,c)
```
